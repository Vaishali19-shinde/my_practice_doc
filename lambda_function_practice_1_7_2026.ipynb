```python
#1.Add two numbers using lambda
add=lambda a,b:a+b
print(add(10,15))
```

```python
#2.find square and cube using lambda
square=lambda x:x**2
cube=lambda x:x**3
print(square(5))
print(cube(5))
```

```python
#3.check positive negative or zero
check=lambda x:'positive' if x>0  else 'negative' if x<0 else 'zero'
print(check(5))
print(check(-5))
print(check(0))
```

```python
#4.Filter worsd starting with a specific letter
words=['apple','banana','cherry','stroberry']
specific_letter='a'
filterd_word=list(filter(lambda x:x.startswith(specific_letter),words))
print(filterd_word)
```

```python
words=['apple','banana','cherry','stroberry']
result=list(filter(lambda x:x.startswith('a'),words))
print(result)
```

```python
#5.convert list string to uppercase
words=['apple','banana','cherry','stroberry']
result=list(map(lambda x:x.upper(),words))
print(result)
```

```python
#6. sort student by marks in descending order
students=[("amit",85),("neha",92),("ravi",78)]
result=sorted(students,key=lambda x:x[1],reverse=True)
print(result)

```

```python
#7.sort employees by name length
employee=["om","raj","kavita","savitaa"]
result=sorted(employee,key=lambda x:len(x))
print(result)
```

```python
#8.find maximum salary employee
employees=[{"name":"amit","salary":450000},
           {"name":"vandana","salary":600000},
           {"name":"raju","salary":890000}]
result=max(employees,key=lambda emp:emp["salary"])
print(result)
```

```python
#9.remove empty string from a string
data=["programming","AI","","machinde learning"]
result=list(filter(lambda x:x!="",data))
print(result)
```

```python
#10.calculate final price after discount
final_price=lambda price,discount:price-(price*discount/100)
print(final_price(1000,10))
```

```python
#11.use lambda with reduce()to find maximum
from functools import reduce
numbers=[12,134,23,13,44]
maximum=reduce(lambda x,y:x if x>y else y,numbers)
print(maximum)
```

```python
#12.lambda vs definition function
def double_def(x):
  return x*2
double_lambda=lambda x: x*2
print(double_def(6))
print(double_lambda(6))
```
