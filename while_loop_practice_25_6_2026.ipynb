#WHILE LOOP

```python
#Basic countiong from 1 to 5
```

```python
i=1
while i<=5:
  print(i)
  i=i+1
```

```python
#reverse counting from 10 to 1

i=10
while i>=1:
  print(i)
  i=i-1
```

```python
#sum of first 10 natural numbers
```

```python
i=1
total=0
while i<=10:
  total=total+i
  i=i+1
print('Total is',total)
```

```python
#multiplication table
num=5
i=1
while i<=10:
  print(num,"*",i,'=',num*i)
  i=i+1
```

```python
#password input until correct
password=""
while password!='python123':
      password=input("enter a password=")
print("Access grantes")
```

```python
# Guess the number game

secret=7
guess=0
while guess!=secret:
  guess=int(input("Enter a guess"))
print('correct guess')
```

```python
#even number from 2 to 20

i=2
while i<=20:
  print(i)
  i=i+2
```

```python
#factorial of number

num=5
fact=1
while num>0:
  fact =fact *num
  num=num-1
print("factorial number",fact)
```

```python
#menu driven program
choice=0
while choice!=5:
    print("1.Add")
    print("2.Substraction")
    print("3.multiplication")
    print("4.division")
    print("5.exit")
    choice = int(input("Enter our choice"))
    if choice==1:
      print("you selected add")
    elif choice==2:
      print("you selected substraction")
    elif choice==3:
      print("you selected multiplication")
    elif choice==4:
      print("you selected division")
    elif choice==5:
      print("thank u")
    else:
      print("Invalid choice")

```

```python
from re import A
#ATM pin with limited access


pin= 1234
attempts=3
while attempts>0:
  enter_pin = int(input("enter a pin="))
  if enter_pin == pin:
    print("login successful")
    break
  else:
    attempts = attempts-1
    print("wrong pin attempts left:",attempts)
  if attempts==0:
     print("card blocked")
```
