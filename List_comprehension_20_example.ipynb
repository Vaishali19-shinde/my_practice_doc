```python
#14.extract first letter in each word

word=['python','Api','Json','Google colab']
first_word=[word[0] for word in word]
print(first_word)
```

```python
#15.extract last letter in each letter
word=['python','Api','Json','Google colab']
last_letter=[word[-1]for word in word]
print(last_letter)
```

```python
#16. search for specific char in given words

word=['python','Api','Json','Google colab']
char_sp=[word for word  in word if word.startswith('p')]
print(char_sp)
```

```python
#17.Remove empt string from a list
x=['python','Api','Json','Google colab','']
rem_list=[x for x in x if x!='']
print(rem_list)
```
