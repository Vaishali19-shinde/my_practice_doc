```python
#program to clone or copy list

# Define the original list
original_list = ["apple", "banana", "cherry"]

# Method 1: Using the .copy() method (Recommended for readability)
cloned_list_1 = original_list.copy()

# Method 2: Using the slicing technique [:] (Fastest execution)
cloned_list_2 = original_list[:]

# Method 3: Using the list() constructor function
cloned_list_3 = list(original_list)

# --- Verification Test ---
# Modify one of the cloned lists to prove it is a separate copy
cloned_list_1.append("orange")

print("Original List:   ", original_list)
print("Cloned List 1:   ", cloned_list_1)
print("Cloned List 2:   ", cloned_list_2)
print("Cloned List 3:   ", cloned_list_3)
```

```python
#11.write a pyhton function that takes two lists and returns true if they have atleat one common
def at_least_two_coomonele(list1,list2):
  return not set(list1).isdisjoint(list2)


# --- Example Usage ---
list_a = [1, 2, 3, 4, 5]
list_b = [5, 6, 7, 8, 9]
list_c = [10, 11, 12]

print(at_least_two_coomonele(list_a, list_b))  # Output: True (shared element: 5)
print(at_least_two_coomonele(list_a, list_c))  # Output: False
```
